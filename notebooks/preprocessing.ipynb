### Step 2: Categorical Encodding

```python
df = pd.read_csv("../data/cleaned/Cleaned_BMW_Sales_2010-2024.csv")

df_categorical = df.select_dtypes(include='object')
df_categorical.nunique().sort_values()
```

Since cardinality is low, we can do OneHotEncoding which used for models

```python
df = pd.read_csv("../data/cleaned/Cleaned_BMW_Sales_2010-2024.csv")

categorical_cols = df.select_dtypes(include='object').columns.tolist()

encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_data = encoder.fit_transform(df[categorical_cols])

encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

df_numeric = df.drop(columns=categorical_cols)

df = pd.concat([df_numeric.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)

df
```

```python
df.to_csv('../data/encoded/Encoded_For_Model_BMW_Sales_2010-2024.csv', index=False)
print("Encoded dataset for model learning saved successfully!")
```
